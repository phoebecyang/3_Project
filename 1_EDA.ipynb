```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
import seaborn as sns

df = pd.read_csv('/Users/phoebeyang/Downloads/data_jobs.csv')
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
import ast

df['job_skills'] = df['job_skills'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str)
    else x if isinstance(x, list)
    else []
)
```

Roles to Explore

```python
df_plot = df['job_title_short'].value_counts().to_frame()

sns.set_theme(style ='ticks')
sns.barplot(data=df_plot, x='count', y='job_title_short', hue='count',palette='dark:g_r',legend=False)
sns.despine()
plt.xlabel('Count')
plt.ylabel('')
plt.title('Numbers of Job per Job Title')
plt.show()
```

Countries to explore

```python
df_countries = df['job_country'].value_counts().head(20).to_frame()
sns.set_theme(style = 'ticks')
sns.barplot(data=df_countries, x='count', y='job_country', hue='count',palette='dark:b_r',legend=False)
sns.despine()
plt.xlabel('Count')
plt.ylabel('')
plt.title('Numbers of Job per Country')
plt.show()
```

Companies to explore

```python
df_copanies = df['company_name'].value_counts().head(20).to_frame()[1:]
sns.set_theme(style = 'ticks')
sns.barplot(data=df_copanies, x='count', y='company_name', hue='count',palette='dark:r_r',legend=False)
sns.despine()
plt.xlabel('Count')
plt.ylabel('')
plt.title('Numbers of Job per Company')
plt.show()
```

Job opportunities

```python
dict_column ={
    'job_work_from_home' : 'Work from Home Offered',
    'job_no_degree_mention' : 'Degree Requirement',
    'job_health_insurance' : 'Health Insurance Offered'
}
fig,ax = plt.subplots(1,3, figsize=(11,13.5))

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), labels=['True', 'False'], autopct='%1.1f%%', startangle=90, explode=(0, 0.1))
    ax[i].set_title(title)
plt.show()
```

Locations for data analysts in the United States

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')]
df_plot = df_DA_US['job_location'].value_counts().head(15).to_frame()
sns.set_theme(style = 'ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue= 'count',palette='dark:b_r',legend=False)
sns.despine()
plt.xlabel('Count')
plt.ylabel('')
plt.title('Numbers of Data Analyst Job per Location in the US')
plt.show()
```

Job opportunities for data analysts in United States

```python
dict_column = {
    'job_work_from_home' : 'Work from Home Offered',
    'job_no_degree_mention' : 'Degree Requirement',
    'job_health_insurance' : 'Health Insurance Offered'
}
fig,ax = plt.subplots(1,3, figsize=(11,13.5))

for i, (column,title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(), labels=['True','False'], autopct='%1.1f%%', startangle=90, explode=(0, 0.1))
    ax[i].set_title(title)
plt.show()
```

Top companies for data analysts in United States

```python
df_companies = df_DA_US['company_name'].value_counts().head(15).to_frame()
sns.set_theme(style = 'ticks')
sns.barplot(data=df_companies, x = 'count', y = 'company_name', hue = 'count', palette='dark:r_r', legend=False)
sns.despine()
plt.xlabel('Count')
plt.ylabel('')
plt.title('Numbers of Data Analyst Job per Company in the US')
plt.show()
```
